```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in ob:
    if i!= 'Is delivering now':
        df[i]=le.fit_transform(df[i])
        
```

```python
df.head()
```

```python
df['Is delivering now'].value_counts()
```

```python
df['Is delivering now'] = df['Is delivering now'].map({'No':0,'Yes':1})
```

```python
df.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables = df[['Country Code', 'City', 'Average Cost for two', 'Currency',
       'Has Table booking', 'Has Online delivery', 'Is delivering now',
       'Switch to order menu', 'Price range', 'Aggregate rating',
       'Rating color', 'Rating text', 'Votes']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
vif
```

```python
df.drop(['Currency','Switch to order menu'],axis = 1,inplace = True)
```

```python
plt.figure(figsize = (8,8))
sns.heatmap(df.corr(),annot = True,vmin = -1,vmax=1)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
X = df.drop(['Is delivering now'], axis=1)
y = df[['Is delivering now']]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
lr = LogisticRegression(C=2.0,max_iter=200,)
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(y_pred)
```

```python
y_pred_train  = lr.predict(X_train)
y_pred_train
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_train,y_pred_train))
```
