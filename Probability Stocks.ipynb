```python
import pandas as pd
```

```python
bem_df =pd.read_csv('BEML.csv')
```

```python
gla_df=pd.read_csv('GLAXO.csv')
```

```python
bem_df.head()
```

```python
bem_df = bem_df[['Date','Close']]
bem_df
```

```python
gla_df.head()
```

```python
gla_df = gla_df[['Date','Close']]
gla_df
```

```python
bem_df['Gain'] = bem_df['Close'].pct_change(periods = 1)
gla_df['Gain'] = gla_df['Close'].pct_change(periods = 1)
```

```python
bem_df
gla_df
```

```python
bem_df.isna().sum()
```

```python
gla_df.isna().sum()
```

```python
bem_df = bem_df.dropna()
gla_df = gla_df.dropna()
```

```python
gla_df = gla_df.set_index(pd.DatetimeIndex(gla_df['Date']))
bem_df = bem_df.set_index(pd.DatetimeIndex(bem_df['Date']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
plt.figure(figsize = (10,5))
plt.plot(bem_df.index,bem_df.Close)
plt.xlabel('Time')
plt.ylabel('Close')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(gla_df.index,gla_df.Close)
plt.xlabel('Time')
plt.ylabel('Close')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(gla_df.index,gla_df.Close)
plt.plot(bem_df.index,bem_df.Close)
plt.xlabel('Time')
plt.ylabel('Close')
plt.show()
```

```python
plt.figure(figsize= (10,5))
plt.plot(bem_df['Gain'])
plt.xlabel('Time')
plt.ylabel('Gain')
plt.show()
```

```python
sns.distplot(gla_df.Gain, label = 'Glaxo');
plt.xlabel('gain');
plt.ylabel('Density');
plt.legend();
```

```python
sns.displot(bem_df.Gain,label = 'Gain');
plt.xlabel('Time');
plt.ylabel('Desinity');
plt.legend();
```

```python
sns.distplot(bem_df.Gain, label = 'BEML');
plt.xlabel('gain');
plt.ylabel('Density');
plt.legend();
```

```python
print(bem_df['Gain'].mean()*100)
print(bem_df['Gain'].std())
```

```python
print(gla_df['Gain'].mean()*100)
print(gla_df['Gain'].std())
```

```python
from scipy.stats import norm
nobem = norm(bem_df['Gain'].mean(),bem_df['Gain'].std())
```

```python
nogla = norm(gla_df['Gain'].mean(),gla_df['Gain'].std())
```

```python
# probability making 2 % loss  or higher
nobem.cdf(-0.02)
```

```python
# probability making 2 % loss or higher loss in glaxo 
1-nogla.cdf(-0.02)
```
